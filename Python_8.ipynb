# lambda / anonymous function:
- A small anonymus function.
- It user the keyword 'lambda'.
- It has no return value;infact if retruns a function.
- It is a oneliner.
- It is not reusable.
- It can take multiple arguments but can only have only one expression.

- Syntax:
         lambda x,y:x+y

```python
a=lambda x,y:x**y
print(a(5,2))
```

```python
a= lambda x:'even' if x%2==0 else 'odd'
print(a(9))
```

```python
add=lambda x,y=15,z=24:x+y+z
```

```python
print(add(1))
```

```python
print(add(1,2))
```

```python
print(add(1,2,3))
```

```python
print(add(z=8,y=2,x=1))
```

# Filter:

```python
filter??
```

```python
l1=[10,40,56,27,13,15,70]
divide=list(filter(lambda x:x%4==0,l1))
print(divide)
```

```python
fruits=['apple','banana','cherry','apricot']
s=list(filter(lambda x: x.startswith('a'),fruits))
```

```python
print(s)
```

```python
'apple'.startswith('a')
```

```python
s=list(map(lambda x:x.startswith('a'),fruits))
print(s)
```

# reduce:

```python
from functools import reduce
```

```python
l=[2,5,10,6,4,2]
product=reduce(lambda x,y:x*y,l)
```

```python
print(product)
```

```python
maxvalue=reduce(lambda x,y: x if x>y else y,[23,11,45,10,1])
print(maxvalue)
```
