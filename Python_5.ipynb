# using key in sorted:

```python
def getlast(s):
    return s[len(s)-1]
```

```python
m=['one','two','three','four','five']
```

```python
ascode=sorted(m,key=getlast)
print(ascode)#sorting from last value bcz s[len(s)-1]
```

# WAP to return a list of even position numbers from a given list:
- i/p = [-5,-8,-4,0,2,3,4,6,7,9]
- o/p = [2,4,6]

```python
s=eval(input("enter list: "))
s1=[]
for i in s:
    if i%2==0 and i>0:
        s1.append(i)
    else:
        continue
print(s1)
```

```python
l1=eval(input("enter list: "))
l=[i for i in l1 if i%2==0 and i>0]
print(l)
```
