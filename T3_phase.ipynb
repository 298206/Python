```python
#WAP to take name and age of customer and write it into the file:

name=input("Enter name:- ")
age=input("Enter age:- ")
file=open("customer.txt",'w')
file.write("Customer name:\n")
file.write(name)
file.write("Customer age:\n")
file.write(age)
file.close()
```

# Context manager-with

```python
with open("cust.txt",'w') as f:
    f.write(input("Name: ")+'\n')
    f.write(input("age: "))
```
