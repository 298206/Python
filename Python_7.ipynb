# Dictionary Comprehension:

- It is a concise way to create a dictionary based on certain condition in elegant way.
- It consists of a key value pair followed by the for and if condition within curly braces.

 **Syntax**
 - {iterator:expression for iterator in iterable if condition}

```python
squares={x:x*x for x in range(6)}
print(squares)
```

# using if condition:

```python
d={x:x*x for x in range(11) if x%2==0}
print(d)
```

```python
distances={'delhi':1000,'ahmedabad':5000,'surat':9000}
print({key:value*0.62 for (key,value) in distances.items()})
```

```python
print({key:value*0.62 for (key,value) in distances.items() if value>1000})
```

```python
# WAP for a given sentence-the quick brown box jumps over the lazy dog create the frequency mapping of words.
# i/p = the quick brown box jumps over the lazy dog
#o/p = {'the':2,'quick':1,---}
s="the quick brown box jumps over the lazy dog"
a=s.split(" ")
l={}
for i in a:
    count=a.count(i)
    l[i]=count
print(l)
```
