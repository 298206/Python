```python
#WAP to find the sum of postive and negative numbers using lambda function:
# i/p = [-4,-2,0,1,2,3]   o/p = sopn=6 and sonn=-6
l=[-4,-2,0,1,2,3]
postive=sum(list(filter(lambda x:x>0,l)))
negative=sum(list(filter(lambda x:x<0,l)))
print(postive)
print(negative)
```
