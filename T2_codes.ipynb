# WAP there is only five digits number n such that every one of following ten numbers share exactly one digit in common in the sameposition as n.find that number
i/p =01265,12171,23257,34548,45970,56236,67324,78084,89872,99414


```python
def common(a,b):
    count=[]
    for i in range(5):
        if a[i]==b[i]:
            count.append(a[i])
    return len(count)
l=['01265','12171','23257','34548','45970','56236','67324','78084','89872','99414']

for i in range(10000,100000):
    for x in l:
        if common(str(i),x)!=1:
            break
    else:
        print(i)
            

```

# WAP given a list scores by students return average score of each student's five scores in order.each entry item[1] has item[1][0] the student id,item[1][1] the student score.find average score as follows: 
i/p = [[1,91],[1,92],[2,93],[2,97],[1,60],[2,77],[1,65],[1,87],[1,100],[2,100],[2,76]]
o/p = [[1,87],[2,88]]
- the average score of student 1 is 87
- the average student of student 2 is 88.6,but with integer division ,it is 88.

```python
i1=[[1,91],[1,92],[2,93],[2,97],[1,60],[2,77],[1,65],[1,87],[1,100],[2,100],[2,76]]
list1=[]
list2=[]
for i in i1:
    if i[0]==1:
        list1.append(i[1])
    if i[0]==2:
        list2.append(i[1])
list1=sorted(list1,reverse=True)
list2=sorted(list2,reverse=True)
#print(l1)
score=[]
sum1=0
sum2=0
for i in range(5):
    if i<=5:
        sum1=sum1+list1[i]
    i+=1
for j in range(5):
    if j<=5:
        sum2=sum2+list2[j]
    j+=1
avg1=sum1//5
avg2=sum2//5
score.append([1,avg1])
score.append([2,avg2])
print(score)
```

# WAP given a string that consists of only 0,1,2s,count the number of substrings that have equal number of 0s,1s,2s.
- i/p =102100211
- o/p = combinations=102,021,210,210021,021 so total substrings=5

```python
def count_equal_substrings(s):
    n = len(s)
    total_count = 0
    valid_substrings = []

    for i in range(n):
        count_0 = 0
        count_1 = 0
        count_2 = 0
        for j in range(i, n):
            char = s[j]
            if char == '0':
                count_0 += 1
            elif char == '1':
                count_1 += 1
            elif char == '2':
                count_2 += 1
            
            if count_0 == count_1 and count_1 == count_2 and count_0 > 0:
                total_count += 1
                valid_substrings.append(s[i:j+1])
    
    return valid_substrings, total_count

input_string = "102100211"
combinations, total_substrings = count_equal_substrings(input_string)

print(f"i/p = {input_string}")
print(f"o/p = combinations={', '.join(combinations)} so total substrings={total_substrings}")





```
