# SEEK method:--
- It sets the file cursor's poision at the offset given .

**SYNTAX**
```python
fileobject.seek(offset[,whence])
```
offset=This is the poision of the read/write pointer by default,
- 0 indicates that the poision of the file is abosultate
- 1 means that the poision is relative to the current poision
- 2 means that the poision is relative to the file's end

````python
#WAP to read a text file and print only word starting with 'i' in reverse order:--
# india is my country  o/p= si aidni

file=open("i1.txt","w")
file.write("india is my country.")
file.close()
````

````python
with open("i1.txt") as f1:
    lines=f1.readlines()
    s1=""
    for line in lines:
        for word in line.split():
            if word[0] in 'Ii':
                s1+=word+" "
    print(s1[::-1])
````

````python
with open("sample1.txt") as f,open("sample12.txt","w") as f1:
    for i in f:
        flag=True
        for j in i.split():
            if j[0]=="#" or flag!=True:
                flag=False
                continue
            f1.write(j+" ")
        if flag!=True:
            f1.write("\n")
f.close()
f1.close()
````

````python
#WAP to count the words in a file that are ending with 'e':--
with open("sample1.txt") as f:
    lines=f.readlines()
    count=0
    for line in lines:
        for word in line.split():
            if word.endswith("e"):
                count+=1
    print(count)
f.close()
````

````python
#WAP to exact a list of 4 letter words that starts ends with same letter from given text file:-
'''
i/p: this is hiih 1221 file for test
o/p: hiih 1221 test
'''
file=open("letter.txt","w")
file.write("this is hiih 1221 file for test")
file.close()
````

````python
with open("letter.txt") as f:
    lines=f.readlines()
    s1=""
    for line in lines:
        for word in line.split():
            if len(word)==4 and word[0]==word[-1]:
                s1+=word+" "
    print(s1)
f.close()
````

````python
#WAP to print every next character with # from given file:--
'''
i/p: This is python
o/p: T#h#i#s #i#s# #p#y#t#h#o#n#
'''

file=open("hash.txt","w")
file.write("This is python")
file.close()
````

````python
with open("hash.txt") as f:
    for i in f:
        for j in i:
            print(j,end="#")
f.close()
````
