# GIT BASIC:

# check installation:- 
- git --version

# set up proxy:
- git config --global http.proxy http://192.168.10.251.808
- git config --global https.proxy http://192.168.10.251.808


# Unset and set user.name and user.email:
- to unset:
```
     git config --global --unset-all user.name
     git config --global --unset-all user.email
```  
- to set:
```

        git config --global user.name riyatank
        git config --global user.email rt215@gmail.com

```

     

# Configure default repo name:
- git config --global init.defaultBranch main

# Initalize  the git repo:
- git init

# to add file for staging:
- git add .

# check the status:
- git status

# Commit the file:
- git commit -m "first"

# Date time:

````python
from datetime import date
````

# create date object:

````python
d=date(2025,10,25)
````

````python
print(d)
````

````python
print(d.ctime())
print(d.day)
print(d.max)
print(d.min)
print(d.month)
print(d.resolution)
print(d.today())
print(d.year)
````

# valid leap year:

````python
from datetime import timedelta
d=date(1900,2,28)
nextday=d.replace(day=28)+timedelta(days=1)
print(nextday)
````
