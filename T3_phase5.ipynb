# Constructor Method:
- These are the methods that are useful for any kind of intiallization one wants to perform with the object.
- A constructor is excuted as soon as an object is instantated.

````python
class Mynum:
    def __init__(self):
        print("Calling the init.")
        self.val=0
    def increment(self):
        self.val=self.val+1
        print(self.val)
````

````python
x=Mynum()
````

````python
x.val
````

````python
x.increment()
````

````python
x.val
````

# Example to demonstrate difference between abstraction and encapsulation:

````python
class Enc:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a*self.b
        
````

````python
Enc_object=Enc(3,4)
````

````python
Enc_object.add()
````

## A complete example:

````python
class Employee:
    empcount=0
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        Employee.empcount+=1
    def displaycount(self):
        print(f"Total employee are {Employee.empcount}")
    def displayemployee(self):
        print(f"Name is {self.name} and salary is {self.salary}")
    def __str__(self):
        return f"The employee name is {self.name}"
````

````python
p=Employee("Python",1000)
````

````python
p.displaycount()
````

````python
p.displayemployee()
````

````python
print(Employee.empcount)
````

````python
j=Employee("java",200)
````

````python
j.displaycount()
````

````python
j.displayemployee()
````

````python
print(Employee.empcount)
````

````python
print(j)
````

# Destrucot mathod:
- The del method is known as destructor method in python.It is called when references to the object have been deleted i.e the object is garbage collector.

````python
class Employee:
    def __init__(self):
        print("Employee created")
    def __del__(self):
        print("destructor called,employee deleted")
````

````python
e=Employee()
````

````python
del e
````

# WAP to create class car. create two objects Maruti with red color and price to 10 lakhs and Kia with blue color and price 7 laks.


````python
class Car:
    def __init__(self,color,price,name):
        self.color=color
        self.price=price
        self.name=name
    def __str__(self):
        return f"The car name  is {self.name} and color is {self.color} and price is {self.price}"
````

````python
maruti=Car("red",1000000,"Maruti")
````

````python
Kia=Car("blue",700000,"Kia")
````

````python
print(maruti)
````

````python
print(Kia)
````
