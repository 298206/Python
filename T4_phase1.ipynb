# Super() method:
- If we write constructor in both parent and child class, the parent class constructor is not available child class.
- super() method is used to invoke parent class construcor from the child class.

````python
class Father:
    def __init__(self):
        self.money=10000
        print("Father class constructor.")
    def show(self):
        print("Father class method.")
class Son(Father):
    def __init__(self):
        super().__init__()
        self.money=2000
        self.car="BMW"
        print("Son class constructor.")
    def show(self):
        super().show()
        print("Son class method.")
````

````python
s=Son()
````

````python
s.car
````

````python
f=Father()
````

````python
s.show()
````

# Abstract class:
- A class derived from ABC class belonging to abc module is known as abstract class in python.
```python
from abc import ABC,abstract method
```
# Rules of abstract class:
- One cannot create objects of abstract class
- It is not necessary to declare all methods abstract in abstract class.
- Abstract class can have abstract and concreate methods.
- If the method is abstract,then the class must be abstract.
- The abstract method of abstarct class must be defind in the child classes.

````python
from abc import ABC,abstractmethod
class Defence(ABC):
    def __init__(self):
        self.id=101
        
    @abstractmethod
    def area(self):
        pass#ABSTRACT METHOD
    
    def gun(self):
        print("The gun is AK47")#cONCRIT METHOD
````

````python
class Army(Defence):
    def area(self):
        print("the area is: ",self.id)
    def new(self):
        print("A new method")
````

````python
a=Army()
````

````python
a.area()
````

````python
a.gun()
````

````python
a.new()
````

````python
class Navy(Defence) :
    def area(self):
        print("The area is: ",self.id)
````

````python
n=Navy()
````

````python
n.area()
````

# WAP and create an abstract class Shape with abstract method calculate_area The derivced classes are Rectangle and Circle. Calculate the corresponding areas.

````python
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self,length,width):
        pass
````

````python
class Rec(Shape):
    def area(self,length,width):
        a=length*width
        print("Area for Rec: ",a)
````

````python
import math
class Circle(Shape):
    def area(self,r):
        a=math.pi*r*r
        print("Area for Rec: ",a)
````

````python
r=Rec()
````

````python
r.area(20,20)
````

````python
c=Circle()
````

````python
c.area(2)
````
