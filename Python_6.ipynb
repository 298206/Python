# WAP given a list of postive integer,let as be the sum of digits of the minimal element of A return 0 if s is odd,otherwise return 1:-

```python
l=eval(input("Enter list: "))
l.sort()
m=l[0]
s=0
while m>0:
    digit=m%10
    s+=digit
    m//=10
if s%2==1:
    print(0)
else:
    print(1)
        
```
