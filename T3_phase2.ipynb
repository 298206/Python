```python
#WAP to find the longest word in given file:-
'''
This is file with the longest word.
o/p = longest
'''

file=open("long.txt","w")
file.write("This is file with the longest word.")
file.close()
```

```python
with open("long.txt") as f:
    longest_word=""
    text_content = f.read()
    words = text_content.split()
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word

print(f"The longest word is: {longest_word}")
print(f"Length: {len(longest_word)}")
```

```python
'''WAP a "paper" programme.your solution should prompt for a filename and display the text file 25 lines at a time ,
pausing each time to ask the user to enter the word "continue",In order to show the next 25 lines or
enter the word "stop" to close the file.
'''

def paper(filename):
    with open(filename,encoding='utf-8') as file:
        line=file.readlines()
        total_lines=len(line)
        start=0
        end=min(25,total_lines)
        while start<total_lines:
            for i in range(start,end):
                print(line[i],end="")
                
            user=input("Enter continue to show next 25 or stop")
            if user.lower()=='stop':
                break
            start+=25
            end=min(start+25,total_lines)
paper("mbox-short")
```
