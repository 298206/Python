# try except statment:
**SYNTAX**
```python
try: 
    statments #Run this main block first
except name1:
    statments #Run this if name1 exception is raised
except (name2,name3):
    statments #Run this if any of name2 or name3 exception occurs
except name4 as var:
    statments #Run this if name4 is raised and assigned instance to var
except:
    statements #Run all other exceptions raised
else:
    statment #Run if no exception is raised during try block
```

# Catching the argument of an Exception:

````python
for i in range(3,-3,-1):
    try:
        print(1.0/i)
    except ZeroDivisionError as err:
        print("Zero division error: ",str(err.args[0]))
````

## WAP to find the longest word contaning a,b,c,d,e,f from the wordlist.txt:

````python
longest=""
with open("wordlist") as r:
    line=r.readlines()
    for i in line:
        if {'a','b','c','d','e','f'}.issubset(set(i)) and len(i.strip())>len(longest):
            longest=i.strip()
print(longest)
````
